# 4. Data Pre-processing

## Missing Values 

```python
df.isna().sum()
```

We got the null values in each column. Its actually less compared to the data.

```python
# Handle the null values

df.dropna(axis=0,inplace=True)
```

```python
df.isna().sum()
```

There is no missing values

## Outlier

```python
print(f'before: {len(df)}')

filtered_entries = np.array([True] * len(df))

for col in numerical:
    zscore = abs(stats.zscore(df[col]))
    filtered_entries = (zscore < 3) & filtered_entries
    
df = df[filtered_entries]

print(f'after: {len(df)}')
```

We clear the outlier using z-score method which is 3 or higher.

```python
df.duplicated().sum()
```

There is no duplicates

## Train Test Split Data

```python
X = df.drop('Attrition', axis=1)
y = df['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## Feature Encoding

```python
# Label Encoding
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Overtime']=le.fit_transform(df['Overtime'])
df['Remote Work']=le.fit_transform(df['Remote Work'])
df['Leadership Opportunities']=le.fit_transform(df['Leadership Opportunities'])
df['Innovation Opportunities']=le.fit_transform(df['Innovation Opportunities'])
```

```python
categorical.columns
```

```python
# One hot encoding
df = pd.get_dummies(df).astype(int)
```

### Logistic regression

```python
## 2. Decision Tree model
```

```python
# Initializing and training the Decision Tree model
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

# Making predictions
y_pred = dt_classifier.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)

# Perform cross-validation
cv_scores = cross_val_score(dt_classifier, X_train, y_train, cv=5)

# Print cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())
```

```python
### 3. Random Forest model
```

```python
# Initializing and training the Random Forest model
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

# Making predictions
y_pred = rf_classifier.predict(X_test)

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)

# Perform cross-validation
cv_scores = cross_val_score(rf_classifier, X_train, y_train, cv=5)

# Print cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", cv_scores.mean())
```
